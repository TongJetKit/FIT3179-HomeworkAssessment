```python
import pandas as pd

# Load the first CSV dataset into a DataFrame
df1 = pd.read_csv('Homework_Week_9/data/WHR_2023.csv')


# Load the second CSV dataset into another DataFrame
df2 = pd.read_csv('countryInfo.csv')


# Merge the two DataFrames based on the common column
merged_df = pd.merge(df1, df2, on='Country', how='inner')
```

```python
merged_df
```

```python
clean_df = merged_df.drop('country',axis = 1)
```

```python
clean_df
```

```python
clean_df.rename(columns={"latitude": "Latitude", "longitude": "Longitude"}, inplace = True)
```

```python
clean_df.to_csv("WHR_2023_clean.csv")
```
